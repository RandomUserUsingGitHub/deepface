```python
import numpy as np

def distance(embedding1, embedding2):
   cosine_similarity = np.dot(embedding1, embedding2) / (np.linalg.norm(embedding1) * np.linalg.norm(embedding2))
   print("Cosine_similarity: ", cosine_similarity)
   cosine_similarity = np.clip(cosine_similarity, -1.0, 1.0)
   angular_distance = np.arccos(cosine_similarity)
   print("Angular distance = ", angular_distance)
   print("Chroma distance = ", 1 - cosine_similarity)
   embeddings = np.array([embedding1, embedding2])
   diffs = (
        embeddings[:, np.newaxis, :] - embeddings[np.newaxis, :, :]
    )  # Shape: (n, n, d)
   distances = np.linalg.norm(diffs, axis=-1)
   print("Euclidean distance = ", np.max(distances))
```
